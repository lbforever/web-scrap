```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time
import pandas as pd
from selenium.webdriver.chrome.service import Service
```

```python
link = 'https://www.flipkart.com/search?q=Laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&as-pos=1&as-type=HISTORY&as-backfill=on%27'
```

```python
service = Service('/Users/xuan/Downloads/chromedriver')
service.start()
```

```python
driver = webdriver.Remote(service.service_url)
driver.get(link)
```

```python
import requests
req = requests.get(link)
```

```python
element = driver.find_element(By.CLASS_NAME, '_1fQZEK')
element
```

```python
from bs4 import BeautifulSoup
import pandas as pd
webpage = req.text
soup = BeautifulSoup(webpage, 'html.parser')

products=[] #List to store name of the product
prices=[] #List to store price of the product
processors=[]
rams=[]
ratings=[]
opss=[]
ssds=[]
displays=[]
warranties=[]
for a in soup.findAll(attrs={'class':'_1fQZEK'}):
    #name=a.find_element(By.CLASS_NAME,'_4rR01T')
    name=a.find('div', attrs={'class':'_4rR01T'})
    price=a.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
    #rating=a.find('div', attrs={'class':'_3LWZlK'})
    processor=a.find_all('li', attrs={'class':'rgWa7D'})[0]
    ram=a.find_all('li', attrs={'class':'rgWa7D'})[1]
    ops=a.find_all('li', attrs={'class':'rgWa7D'})[2]
    ssd=a.find_all('li', attrs={'class':'rgWa7D'})[3]
    display=a.find_all('li', attrs={'class':'rgWa7D'})[4]
    warranty=a.find_all('li', attrs={'class':'rgWa7D'})[5]
    
    products.append(name.text)
    prices.append(price.text)
    #ratings.append(rating.text)
    processors.append(processor.text)
    rams.append(ram.text)
    opss.append(ops.text)
    ssds.append(ssd.text)
    displays.append(display.text)
    warranties.append(warranty.text)
df = pd.DataFrame({'Product Name':products,'Price':prices, 'processor':processors, 'ram':rams, 
                   'ops':opss,'ssd':ssds,'display':displays,'warranties':warranty})
df.to_csv('products.csv', index=False, encoding='utf-8')
df
```
